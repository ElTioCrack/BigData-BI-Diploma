# Bloque 4 — Introducción a Pandas

En este bloque trabajaremos con `pandas`, una biblioteca de Python usada para organizar y analizar datos en forma de tabla.


## Importar Pandas

La biblioteca `pandas` nos permite crear estructuras tipo tabla llamadas `DataFrame`.

En la siguiente línea:
- `import pandas as pd` importa la biblioteca
- `pd` es una abreviatura que se usa para escribir menos código


```python
import pandas as pd
```

## Crear un conjunto de datos

Primero construiremos un diccionario llamado `datos`. Cada clave representa una columna y cada lista representa los valores de esa columna.


```python
datos = {
    'empresa': ['Empresa A', 'Empresa B', 'Empresa C', 'Empresa D'],
    'region': ['La Paz', 'Santa Cruz', 'Cochabamba', 'Tarija'],
    'ventas': [200, 350, 150, 400]
}

print(datos)
```

## Crear un DataFrame

La instrucción `pd.DataFrame(datos)` convierte el diccionario en una tabla.

- `DataFrame` es la estructura principal de `pandas`
- `datos` es la información que queremos convertir en tabla


```python
# Convertimos el diccionario en un DataFrame
df = pd.DataFrame(datos)

# Mostramos la tabla resultante
print(df)
```

## Explorar las primeras filas

La función `head()` muestra las primeras filas del DataFrame.


```python
df.head()
```

## Información del DataFrame

La función `info()` muestra:
- número de filas
- número de columnas
- tipo de dato de cada columna


```python
df.info()
```

## Estadísticas básicas

La función `describe()` calcula estadísticas básicas sobre las columnas numéricas.


```python
df.describe()
```

## Operaciones analíticas simples

Una vez que tenemos los datos en un DataFrame, podemos realizar análisis básicos.


### Promedio de ventas

La función `mean()` calcula el promedio de una columna numérica.


```python
df['ventas'].mean()
```

Interpretación:

El resultado representa el promedio de ventas del conjunto de empresas.


### Empresa con mayor venta

Aquí identificamos la fila donde la variable `ventas` tiene el valor más alto.


```python
df.loc[df['ventas'].idxmax()]
```

Explicación:

- `idxmax()` encuentra la posición del valor máximo
- `loc[]` recupera la fila completa correspondiente a esa posición


### Agrupación por región

La función `groupby()` permite agrupar los datos por una categoría.


```python
df.groupby('region')['ventas'].sum()
```

Explicación:

Aquí agrupamos por región y sumamos las ventas de cada grupo. Este tipo de operación es muy común en análisis organizacional.
