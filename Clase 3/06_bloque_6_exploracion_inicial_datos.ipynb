En este bloque realizaremos una primera exploración del dataset para entender sus variables, tipos de datos, estadísticas y agrupaciones básicas.


## Importar Pandas y cargar el dataset


```python
import pandas as pd

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('ventas_empresas.csv')

# Mostrar las primeras filas
df.head()
```

## Tipos de variables

La instrucción `df.dtypes` muestra el tipo de dato de cada variable.

Interpretación general:
- `object` = texto
- `int` = número entero
- `float` = número decimal


```python
df.dtypes
```

## Estadísticas básicas del dataset

Una vez identificadas las variables, el siguiente paso es obtener estadísticas básicas del dataset.

Estas estadísticas permiten responder preguntas como:
- ¿Cuál es el promedio de ventas?
- ¿Cuál es el valor máximo?
- ¿Existe mucha variabilidad?


```python
df.describe()
```

## Variables categóricas

Las variables categóricas describen grupos o categorías.

Ejemplos:
- región
- sector
- tipo de cliente

Para analizarlas utilizamos `value_counts()`.


```python
df['region'].value_counts()
```

## Análisis de una variable individual


### Promedio de ventas


```python
df['ventas'].mean()
```

### Valor máximo


```python
df['ventas'].max()
```

### Valor mínimo


```python
df['ventas'].min()
```

## Agrupación por categorías

Una operación analítica muy común es agrupar datos por categorías.

Ejemplo: ventas por región.


```python
df.groupby('region')['ventas'].sum()
```

## Ejemplo completo para clase


```python
import pandas as pd

# Cargar dataset
#df = pd.read_csv('ventas_empresas.csv')
df = pd.read_csv('https://raw.githubusercontent.com/ElTioCrack/BigData-BI-Diploma/refs/heads/main/ventas_empresas.csv')

# Mostrar primeras filas
print(df.head())

# Tipos de variables
print(df.dtypes)

# Estadísticas básicas
print(df.describe())

# Distribución por región
print(df['region'].value_counts())

# Promedio de ventas
print(df['ventas'].mean())

# Agrupación por región
print(df.groupby('region')['ventas'].sum())
```
