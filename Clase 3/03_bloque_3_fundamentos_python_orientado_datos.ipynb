# Bloque 3 — Fundamentos de Python orientado a datos

En este bloque veremos los elementos básicos de Python que se utilizan con frecuencia en análisis de datos.


## Variables

Una variable es un nombre que se utiliza para guardar información. Por ejemplo, una variable puede almacenar un precio, un nombre de cliente o un valor lógico.


```python
# Guardamos un número decimal en la variable precio
precio = 50.75

# Guardamos un texto en la variable cliente
cliente = 'Empresa A'

# Guardamos un valor lógico en la variable activo
activo = True

# Mostramos los valores guardados
print(precio)
print(cliente)
print(activo)
```

## Tipos básicos de datos

En este ejemplo aparecen tres tipos de datos:
- `float`: número decimal
- `str`: texto
- `bool`: valor lógico (`True` o `False`)


```python
# La función type() permite conocer el tipo de dato de cada variable
print(type(precio))
print(type(cliente))
print(type(activo))
```

## Listas de datos

En análisis de datos es común trabajar con varios valores a la vez. Python permite almacenarlos en listas.


```python
# Creamos una lista con varias observaciones de ventas
ventas = [100, 120, 150, 200]

# Mostramos la lista completa
print(ventas)
```

## Operaciones básicas con listas

Con una lista podemos realizar cálculos simples como la suma y el promedio.


```python
# Sumamos todos los valores de la lista
print(sum(ventas))

# Calculamos el promedio de la lista
print(sum(ventas) / len(ventas))
```

## Diccionarios

Un diccionario permite almacenar información estructurada mediante pares de clave y valor.


```python
# Creamos un diccionario con información de un cliente
cliente = {
    'nombre': 'Empresa A',
    'ventas': 200,
    'region': 'La Paz'
}

# Mostramos el diccionario
print(cliente)
```

## Introducción al concepto de tabla de datos

En análisis de datos normalmente no trabajamos con una sola lista o un solo diccionario. Trabajamos con tablas donde cada fila representa una observación y cada columna representa una variable.
