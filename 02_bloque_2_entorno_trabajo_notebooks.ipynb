# Bloque 2 — Entorno de trabajo para análisis de datos

En este bloque conoceremos el entorno donde se desarrolla el análisis de datos con Python. La idea es entender dónde se escribe el código, cómo se ejecuta y dónde aparecen los resultados.


## ¿Qué es un notebook?

Un notebook es un documento interactivo que permite combinar:
- texto explicativo
- código ejecutable
- resultados del análisis

Esto permite explicar el proceso y ejecutar el análisis en un mismo lugar.


## Tipos de celdas en un notebook

En un notebook se utilizan principalmente dos tipos de celdas:

**1. Celdas de texto**
Sirven para escribir explicaciones, títulos, pasos del análisis e interpretaciones.

**2. Celdas de código**
Sirven para ejecutar instrucciones en Python y ver resultados, tablas o gráficos.


## Primera celda de código

## ¿Cómo se usa este entorno en clase?

La secuencia habitual es:
1. escribir una explicación en una celda de texto
2. ejecutar una celda de código
3. observar el resultado
4. interpretar lo que ocurrió

Esta lógica se mantiene en todo el proceso de análisis de datos.
