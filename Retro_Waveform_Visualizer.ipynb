```python
import time, random, numpy as np
import tkinter as tk
from threading import Thread

WIDTH, HEIGHT = 800, 400
BG_COLOR, BAR_COLOR = "#000a00", "#00FF41"
NUM_BARS = 30
BAR_WIDTH = WIDTH / NUM_BARS

class ChiptuneVisualizer:
    def __init__(self, root):
        self.root = root
        self.root.title(">> 8-Bit Spectrum Analyzer :: ELTIOCRACK <<")
        self.canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg=BG_COLOR, highlightthickness=0)
        self.canvas.pack()
        
        self.bars = [self.canvas.create_rectangle(i*BAR_WIDTH+2, HEIGHT, (i+1)*BAR_WIDTH-2, HEIGHT, fill=BAR_COLOR, outline=BG_COLOR) for i in range(NUM_BARS)]
        
        self.running = True
        Thread(target=self.animate_loop, daemon=True).start()

    def animate_loop(self):
        base_levels = np.sin(np.linspace(0, 4 * np.pi, NUM_BARS)) * 0.4 + 0.5
        while self.running:
            for i, bar in enumerate(self.bars):
                target_h = (base_levels[i] * random.uniform(0.5, 1.5)) * HEIGHT * 0.8
                coords = self.canvas.coords(bar)
                curr_h = HEIGHT - coords[1]
                new_y = HEIGHT - (curr_h + (target_h - curr_h) * 0.4)
                # Usar after() para que el hilo principal actualice la UI (Thread-safe)
                self.root.after(0, self.canvas.coords, bar, coords[0], new_y, coords[2], HEIGHT)
            time.sleep(0.04)

root = tk.Tk()
app = ChiptuneVisualizer(root)
root.mainloop()
```
