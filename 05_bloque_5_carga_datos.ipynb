# Bloque 5 — Carga de datos para análisis

En este bloque aprenderemos a cargar un dataset desde un archivo CSV y revisar su estructura básica.


## Importar Pandas

Usamos `pandas` porque esta biblioteca permite leer archivos como CSV y trabajar con ellos como tablas.


```python
import pandas as pd
```

## Cargar el dataset

La función `pd.read_csv()` lee un archivo CSV y lo convierte en un DataFrame.

En este caso, el archivo se llama `ventas_empresas.csv`.


```python
# Leemos el archivo CSV y lo guardamos en la variable df
df = pd.read_csv('ventas_empresas.csv')

# Mostramos el contenido cargado
df
```

## Mostrar primeras filas

`head()` permite observar las primeras filas del dataset.


```python
df.head()
```

## Mostrar últimas filas

`tail()` sirve para verificar la parte final del dataset.


```python
df.tail()
```

## Información del dataset

`info()` permite ver la estructura del dataset, incluyendo filas, columnas y tipos de datos.


```python
df.info()
```

## Dimensión del dataset

`shape` devuelve una tupla con:
- número de filas
- número de columnas

Ejemplo: `(1000, 5)` significa 1000 registros y 5 variables.


```python
df.shape
```

## Variables disponibles

`columns` permite ver los nombres de las columnas del dataset.


```python
df.columns
```

## Ejemplo completo

El siguiente ejemplo resume el proceso de carga e inspección básica del dataset.


```python
import pandas as pd

# Cargar dataset
df = pd.read_csv('ventas_empresas.csv')

# Mostrar primeras filas
print('Primeras filas del dataset')
print(df.head())

# Mostrar estructura
print('\nInformación del dataset')
print(df.info())

# Dimensión del dataset
print('\nDimensión del dataset')
print(df.shape)

# Mostrar nombres de columnas
print('\nVariables disponibles')
print(df.columns)
```
